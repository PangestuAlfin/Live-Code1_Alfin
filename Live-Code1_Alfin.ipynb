**a. Buatlah dua fungsi eksplisit masing-masing memuat perhitungan hambatan (R) dan daya (P) yang nantinya akan digunakan untuk perhitungan**

**JAWABAN SOAL A** 
karena di minta buat 2 fungsi yang masing-masing menghitung hambatan dan daya. maka langkah pertama kita harus mengetahui rumus tersebut. jika sudah mengetahui buat masing masing fungsi dengan rumus masing-masing. 

fungsi pertama membuat perhitungan hambatan yang rumusnya : R=V/I

fungsi kedua membuat perhitungan daya yang rumusnya: P= V*I

nnti akan di gunakan untuk perhitungan. brrti fungsi ini akan di pakai berulang kali sesuai jumlah data. maka kita gunakan *return*

```python
#fungsi pertama menghitung Hambatan
def hitung_R(V,I):
  rumus_R = V/I
  return rumus_R

#fungsi kedua menghitung Daya
def hitung_P(V,I):
  rumus_P = V*I
  return rumus_P
```

**b. Buat kolom baru dengan nama kolom R/P berisikan hasil perhitungan daya/hambatan dengan kriteria untuk Kapasitas >=2.5 hitung daya, Kapasitas <2.5 hitung nilai hambatan. (Hint: Gunakan for loop dan if statement untuk mengerjakan perintah nomor b, kamu bisa menggunakan list untuk menyimpan perhitungan sementara, setelahnya dapat dimasukkan ke kolom R/P)**

**Jawaban b**
rumus menghitung daya : Volt * Current/Ampere (P=V*I)

rumur menghitung hambatan : volt/ampere (R=V/I)

V pada kolom = V_m
I pada kolom = Current (A)

karena yang di minta nilai kapasitas yang >=2.5 hitung daya. 
< 2.5 hitung nilai hambatan

maka nilai pada kolom **cap** akan di buatkan kondisi yang mana >= 2.5 dia akan menghitung nilai pada kolom voltage dan nilai dari kolom current. dan apabila nilai yang di masukan lebih keci dari 2.5 dia akan mencetak hitung hambatan 

```python
#r_p disini membuat nilai baru
r_p = []
#for disini membuat kondisi, index, row pada data kita untuk mengulang pada setiap baris dalam dataframe
for i, row in df.iterrows():
  #membuat kolom yang akan di masukan pada statment
    voltage = row['V_m']
    current = row['Current (A)']
    kapasitas = row['Cap']
    #if statment ini untuk lebih besar dari 2.5
    if kapasitas >= 2.5:
        # dengan menghitung fungsi kedua
        hitung_daya = hitung_P(voltage, current)
        #lalu nilai pada baris akan masuk ke menghitung fungsi 
        r_p.append(hitung_daya)
    #else ini statment yang lebih kecil dari statment if di atas
    else:
        #dengan menghitung fungsi pertama
        hitung_hambatan = hitung_R(voltage, current)
         #lalu nilai pada baris akan masuk ke menghitung fungsi 
        r_p.append(hitung_hambatan)
#buat kolom baru 
df['R/P'] = r_p
```

**SOAL C**

Hitunglah rata-rata nilai kapasitas untuk arus = 1 Ampere dan cycle ganjil.

**JAWABAN SOAL C**


```python
#membuat variabel yang diisi dengan statment 
#dataframe pada kolom 'current (A)' == 1 dana pada kolom 'cycle' ganji buat nilai rata-rata dari kolom 'Cap' atau kapasitas 
#untuk memilih bilangan ganjil maka %2 ==1 untuk genap %2 == 0

rata2 = df[(df['Current (A)'] == 1) & (df['Cycle'] % 2 == 1)]['Cap'].mean()
```
